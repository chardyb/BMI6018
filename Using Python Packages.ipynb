```python
#I am loading in my own example dataset that I downloaded from Kaggle. This is the same dataset I used for last week's assignment.
##https://www.kaggle.com/datasets/prasad22/healthcare-dataset?resource=download
df1 = pd.read_csv('/Users/carterhardy/Downloads/healthcare_dataset.csv')
```

```python
df1.head(3)
```

```python
#displot will make histograms of subsets of data, we can see the distribution of age, split up by gender, for every medical condtion 
#in the example data we have.
sns.displot(
    df1, x='Age', col = 'Medical Condition', row='Gender',
        binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
```

```python
#Here we are doing grouped bar plots by using catplot()
#we have the x axis be Gender, splitting it male and female, and then we can see what the billing amount is for each insurance provider split
#up by gender. We dont actually see anything useful because this is example data and it is probably pretty randomly set, but in a real world
#setting we could use this to see if male or females have a higher billing amount based on their insurance provider and this graph would give
#us a great display of what that data would look like.
sns.catplot(
    data=df1, kind="bar",
    x = 'Gender', hue="Insurance Provider", y="Billing Amount", 
    errorbar="sd", palette="dark", alpha=.6, height=6
)
```
