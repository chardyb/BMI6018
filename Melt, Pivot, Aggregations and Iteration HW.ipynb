```python
#Pivot 
```

```python
#I found that using pivot() had a lot of issues if there was any duplicate value, I didnt want to take out any 
#duplicates because if a condition and insurance provider is what we are trying to pivot on, there can be many 
#details that are different even if those are the same. So I made a smaller subset of data to demonstrate pivot()
df_sub = df.head(15)

#here is the subset data that will be used to demonstrate pivot()
df_sub
```

```python
#Our row for pivot() will be Hospital, columns will be the insurance provider, and then the values are medical 
#condition and billing amount.
#From this view, we can see what hospital was visited with what insurance and what medical condition was seen and 
#then we can see how much the trip cost to that hospital with that insurance provider.
df_sub.pivot(index = 'Hospital', columns = 'Insurance Provider', values = ['Medical Condition', 
                                                                           'Billing Amount']).fillna('NA')
```

```python
#I did not love that output. I don't think it looked very good. To be honest, I haven't thought of a way that 
#pivot() would be useful in the context that there can't be a duplicate. I did another example to try to find 
#a better demonastration of pivot()

#In this example, our rows are the insurance provider and the hospital visited, our column in the medical condition
#and our value is how much the bill cost. This is a better way to display this information, so that someone can 
#see where the  visit was with what insurance, what condition was the purpose of the visit, and how much it costs.

#With large datasets it is going to be harder to use pivot() because there will most likely be duplicate conditions
#It is not unreasonable to believe that 2 people with the same insurance would visit the same hospital for the same
#condition. Even more so, it is beleivable that only one of those combinations are true, and in that case pivot()
#would not be useful.
#You could take duplicate values out, but most of the time that would be taking out useful data.
df_sub.pivot(index = ['Insurance Provider', 'Hospital'], columns = 'Medical Condition', 
             values = ['Billing Amount']).fillna('0')

```

```python
#Aggregation 
```

```python
#In this example I aggregated the mean for the billing amount of the data frame and then the median
#max, and min for the age of the patients. This was to display that with agg() you can do different aggregations 
#and they will display together.
#This can be useful information. We see the median age of patients, we can see the oldest and youngest patient age.
#This also will calculate the mean billing amount for a hospital visit.
df.agg({'Billing Amount': ['mean'],
       'Age': ['median', 'max', 'min']}
      )
```

```python
#Jumping a little ahead. If we use groupby(), we can get the insurance provider and then calculate out the 
#mean, count, sum, and max for the billing amount for each insurance provider. This is useful information and 
#an effective way to use agg() and groupby(). Now we can see how many people used what insurance provider

df.groupby("Insurance Provider")['Billing Amount'].agg(['mean','count','sum', 'max'])
```

```python
#I am creating another subset of data to display iterations and so that I do not need to iterate over every row of
#the data frame
df_subset = df.tail(10)
```

```python
#I am interating over every row of the column 'Name' and printing it out
for index, row in df_subset.iterrows():
    print(row['Name'])
```

```python
#Seeing that the case is random, I am now interating over every row to correct to title case, so that the first 
#letter of both names will be capitalized and the rest of the letters will be lowercase. 
for index, row in df_subset.iterrows():
    df_subset.at[index, 'Name'] = row['Name'].title()
```

```python
#I will show just the column 'Name' to show that each name is now in title case. This is to save space and because
#I don't want to print the entire subset of the dataframe, I just want to focus on the names to display the 
#change that has occured.
df_subset['Name']
```

```python
#Groupby
```

```python
#Using groupby, I can group the insurance providers together and then display each provider's mean billing amount

df.groupby('Insurance Provider')['Billing Amount'].mean()
```

```python
#Going back and adding the agg() function, we can group each insurance provider and display the count, mean, 
#and median of the billing amount of each one.

df.groupby('Insurance Provider')['Billing Amount'].agg(['count','mean', 'median'])
```

```python
#To make this more complex, we can add the medical condition into the groupby() as well.
#This is a great way for us to see some simple stats quickly. We can see how many patiets from an insurance company
#are being seen for a specific medical condition and how much the billing amount might be based on the mean or 
#median for that medical condition under that insurance.

df.groupby(['Insurance Provider', 'Medical Condition'])['Billing Amount'].agg(['count','mean', 'median'])
             
```

```python
#While doing research and watching youtube videos on groupby() I came across this interesting use where you group
#on a key and frequency by using pd.Grouper. To display this, I am going to change the column 'Data of Admission'
#to a datatime object so I can use it as the key to group together.

df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])

#Next, using pd.Grouper, we are going to use the frequency of 1 month to group the 'Date of Admission' into months
#and then show the sum of the billing amount. Using the tail(12) function, we can see the sum of the billing amount
#for the last 12 months of data we have.

df.groupby(pd.Grouper(key='Date of Admission', freq = '1m'))['Billing Amount'].sum().tail(12)
```
